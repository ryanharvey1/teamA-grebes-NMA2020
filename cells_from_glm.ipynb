```
import matplotlib.pyplot as plt
plt.style.use('dark_background')
params = {'legend.fontsize': 'large',
         'axes.labelsize': 'large',
         'axes.titlesize':'large',
         'xtick.labelsize':'large',
         'ytick.labelsize':'large'}
plt.rcParams.update(params)

%config InlineBackend.figure_format = 'retina'
import pandas as pd
import numpy as np
import os, requests
from google.colab import files
import math
```

```
print('loading speed tuning data...')
speed_map = np.load('/content/drive/My Drive/nma_data/6163622_v6/speed_map.npy')
df = pd.read_pickle("/content/drive/My Drive/nma_data/6163622_v6/speed_df.pkl")
idx_HTN = np.load("/content/drive/My Drive/nma_data/6163622_v6/run_onset/idx_319.npy")
```

```
# idx = np.array(range(len(speed_map)))
# idx = idx[idx_HTN]
idx = idx_HTN
```

```
rc = math.ceil(np.sqrt(len(idx)))
fig, axs = plt.subplots(rc,rc, figsize=(20, 20), edgecolor='k')
fig.subplots_adjust(hspace = 0.5, wspace=0.5)

axs = axs.ravel()

for i in range(len(idx)):
  axs[i].plot(speed_map[idx[i]],linewidth=3)
  axs[i].axis('off')
# fig.savefig("glm_identified_tuning.png",dpi=300)
# files.download("glm_identified_tuning.png")  
```

```
plt.scatter(df.speed_mod,df.speed_cor,label='all cells')
plt.scatter(df.speed_mod[idx],df.speed_cor[idx],label='GLM identified cells')
plt.xlabel('speed modulation')
plt.ylabel('speed correlation')
plt.legend()

# plt.savefig("glm_identified_scatter.png",dpi=300)
# files.download("glm_identified_scatter.png")  
```
